```python
#Задание 1
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv("HR.csv")
df.head()
```

```python
#Задание 2
df.describe()
```

```python
#Задание 3
metrics = ["satisfaction_level", "last_evaluation", "number_project", "average_montly_hours", "time_spend_company"]
min_corr = [1,"",""]
max_corr = [0,"",""]

for i in metrics:
  for j in metrics:
    if i==j: continue
    df.plot(kind="scatter", x=i, y=j)
    corr_p = df[[i, j]].corr('pearson')
    print("Коэффициент корреляции Пирсона\n", corr_p)
    plt.show()
    abs_corr_p = abs(corr_p.to_numpy()[0][1])
    if min_corr[0] > abs_corr_p:
      min_corr[0] = abs_corr_p
      min_corr[1] = i
      min_corr[2] = j
    if max_corr[0] < abs_corr_p:
      max_corr[0] = abs_corr_p
      max_corr[1] = i
      max_corr[2] = j

print("Наименее скоррелированные переменные:", min_corr)
print("Наиболее скоррелированные переменные:", max_corr)
```

```python
#Задание 7
from scipy import stats

alpha = 0.05
a = df[df["salary"]=="high"]
A = a["average_montly_hours"]
b = df[df["salary"]=="low"]
B = b["average_montly_hours"]

p_val = stats.ttest_ind(A,B).pvalue
if p_val < alpha: 
  print('Время проведенное на работе значимо влияет на зарплату:', p_val)
else:
  print('Время проведенное на работе значимо не влияет на зарплату:', p_val)
```

```python
#Задание 8
def stat(left):
  df2 = df[df["left"]==left]
  print("Доля сотрудников с повышением за последние 5 лет, %:", df2["promotion_last_5years"].mean()*100)
  print("Средняя степень удовлетворенности (max=1.0):", df2["satisfaction_level"].mean())
  print("Среднее количество проектов:", df2["number_project"].mean())

print("Среди уволившихся:")
stat(True)

print("Среди не уволившихся:")
stat(False)
```

```python
#Задание 9
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

x = df[["satisfaction_level","last_evaluation","number_project","average_montly_hours","time_spend_company","Work_accident","promotion_last_5years"]]
y = df["left"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
result = pd.DataFrame([y_test, lda.predict(x_test)]).T
result.head(50)
```
